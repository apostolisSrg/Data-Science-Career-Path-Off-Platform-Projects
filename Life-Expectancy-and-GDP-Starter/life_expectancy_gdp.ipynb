```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

life_expectancy_gdp = pd.read_csv("all_data.csv")

formatted_numbers_life_exp_GDP = []
for number in life_expectancy_gdp.GDP:
    number = int(number)
    formatted_number = f"{number:,}".replace(",", ".")
    formatted_numbers_life_exp_GDP.append(formatted_number)

life_expectancy_gdp_series = pd.Series(formatted_numbers_life_exp_GDP, name="GDP")

print(f"\n{life_expectancy_gdp_series.head(50)}")
life_expectancy_gdp.head()
```

```python
plt.figure(figsize=(17, 10), facecolor='#F0F8FF')


ax1 = plt.subplot(2, 3, 1)
plt.plot(life_expectancy_gdp[life_expectancy_gdp.Country == "Chile"].Year, life_expectancy_gdp[life_expectancy_gdp.Country == "Chile"]["Life expectancy at birth (years)"], color='red',marker="*")
ax1.set_xlabel("Year")
ax1.set_ylabel("Life Expectancy (Years)")
ax1.legend(["Chile"])
ax1.grid(True)
ax1.set_facecolor('#FFF8DC')

ax2 = plt.subplot(2, 3, 2)
plt.plot(life_expectancy_gdp[life_expectancy_gdp.Country == "Germany"].Year, life_expectancy_gdp[life_expectancy_gdp.Country == "Germany"]["Life expectancy at birth (years)"], color="yellow", marker="v")
ax2.set_xlabel("Year")
ax2.set_ylabel("Life Expectancy (Years)")
ax2.legend(["Germany"])
ax2.grid(True)
ax2.set_facecolor('#FFF8DC')

ax3 = plt.subplot(2, 3, 3)
plt.plot(life_expectancy_gdp[life_expectancy_gdp.Country == "Mexico"].Year, life_expectancy_gdp[life_expectancy_gdp.Country == "Mexico"]["Life expectancy at birth (years)"], color="green", marker="o")
ax3.set_xlabel("Year")
ax3.set_ylabel("Life Expectancy (Years)")
ax3.legend(["Mexico"])
ax3.grid(True)
ax3.set_facecolor('#FFF8DC')

ax4 = plt.subplot(2, 3, 4)
plt.plot(life_expectancy_gdp[life_expectancy_gdp.Country == "United States of America"].Year, life_expectancy_gdp[life_expectancy_gdp.Country == "United States of America"]["Life expectancy at birth (years)"], color="blue", marker="1")
ax4.set_xlabel("Year")
ax4.set_ylabel("Life Expectancy (Years)")
ax4.legend(["United States of America"])
ax4.grid(True)
ax4.set_facecolor('#FFF8DC')

ax5 = plt.subplot(2, 3, 5)
plt.plot(life_expectancy_gdp[life_expectancy_gdp.Country == "Zimbabwe"].Year, life_expectancy_gdp[life_expectancy_gdp.Country == "Zimbabwe"]["Life expectancy at birth (years)"], color="purple", marker="3")
ax5.set_xlabel("Year")
ax5.set_ylabel("Life Expectancy (Years)")
ax5.legend(["Zimbabwe"])
ax5.grid(True)
ax5.set_facecolor('#FFF8DC')

ax6 = plt.subplot(2, 3, 6)
plt.plot(life_expectancy_gdp[life_expectancy_gdp.Country == "China"].Year, life_expectancy_gdp[life_expectancy_gdp.Country == "China"]["Life expectancy at birth (years)"], color="orange", marker="p")
ax6.set_xlabel("Year")
ax6.set_ylabel("Life Expectancy (Years)")
ax6.legend(["China"])
ax6.grid(True)
ax6.set_facecolor('#FFF8DC')

plt.show()
plt.clf()
```
